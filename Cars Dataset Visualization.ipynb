###  Question 2
Create bar plots for the following, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced.

1. Cylinders and Horsepower, Cylinders and miles per gallon, Cylinders and weight.

2. Year and time-to-60, Year and miles per gallon, Year and horsepower.

3. Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.

4. Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.

```python
#Cylinders and Horsepower, Cylinders and miles per gallon, Cylinders and weight.

plt.figure(figsize=(20,10))  # set overall figure size

plt.subplot(1, 3, 1)
sns.barplot(x="cylinders", y="hp", data=df)
plt.title("Cylinders vs HP")

plt.subplot(1, 3, 2)
sns.barplot(x="cylinders", y="mpg", data=df)
plt.title("Cylinders vs MPG")

plt.subplot(1, 3, 3)
sns.barplot(x="cylinders", y="weightlbs", data=df)
plt.title("Cylinders vs Weightlbs")

plt.tight_layout()
plt.show()
```

```python
#Year and time-to-60, Year and miles per gallon, Year and horsepower.
plt.figure(figsize=(25,10))  # set overall figure size

plt.subplot(1, 3, 1)
sns.barplot(x="year", y="hp", data=df)
plt.title("Year vs HP")

plt.subplot(1, 3, 2)
sns.barplot(x="year", y="mpg", data=df)
plt.title("Years vs MPG")

plt.subplot(1, 3, 3)
sns.barplot(x="year", y="time-to-60", data=df)
plt.title("Year vs time-to-60")


plt.show()
```

```python
#Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands.
plt.figure(figsize=(25,10))  # set overall figure size

plt.subplot(2, 2, 1)
sns.barplot(x="cylinders", y="hp", data=df , hue ="brand")
plt.title("cylinders vs HP")

plt.subplot(2, 2, 2)
sns.barplot(x="cylinders", y="mpg", data=df ,hue ="brand")
plt.title("cylinders vs MPG")

plt.subplot(2, 2, 3)
sns.barplot(x="cylinders", y="time-to-60", data=df ,hue ="brand")
plt.title("cylinders vs time-to-60")

plt.subplot(2, 2, 4)
sns.barplot(x="cylinders", y="time-to-60", data=df ,hue ="brand")
plt.title("cylinders vs weightlbs")

plt.show()
```

```python
#Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.
plt.figure(figsize=(18,12))  # set overall figure size

plt.subplot(2, 2, 1)
sns.barplot(x="year", y="hp", data=df , hue ="brand")
plt.title("yearvs HP")
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
sns.barplot(x="year", y="mpg", data=df ,hue ="brand")
plt.title("year vs MPG")
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
sns.barplot(x="year", y="time-to-60", data=df ,hue ="brand")
plt.title("year vs time-to-60")
plt.xticks(rotation=45)

plt.subplot(2, 2, 4)
sns.barplot(x="year", y="time-to-60", data=df ,hue ="brand")
plt.title("year vs weightlbs")
plt.xticks(rotation=45)

plt.show()
```

### # Question 6: Create a histogram for the following:

Each column in the data to understand the distribution of the data.
Create histograms for hp, mpg, time-to-60, weight lbs, cubic inches, year, with respect to brand.

```python
plt.subplots(2,3 , figsize=(18,12))

plt.subplot(2,3,1)
sns.histplot(data =df , x ="hp" , hue = "brand")
plt.title("Brand-wise  HP Distribution")

plt.subplot(2,3,2)
sns.histplot(data =df , x ="mpg" , hue = "brand")
plt.title("Brand-wise  MPG Distribution")

plt.subplot(2,3,3)
sns.histplot(data =df , x ="time-to-60" , hue = "brand")
plt.title("Brand-wise  time-to-60 Distribution")

plt.subplot(2,3,4)
sns.histplot(data =df , x ="weightlbs" , hue = "brand")
plt.title("Brand-wise  Hweight lbs Distribution")

plt.subplot(2,3,5)
sns.histplot(data =df , x ="cubicinches" , hue = "brand")
plt.title("Brand-wise  cubicinches Distribution")

plt.subplot(2,3,6)
sns.histplot(data =df , x ="year" , hue = "brand")
plt.title("Brand-wise year Distribution")

plt.tight_layout()
plt.show()
```

### # Question 7: Create line plots for the following:

Create a Line plot between hp-mpg, hp-(time-to-60), mpg-(time-to-60), hp-weight lbs, mpg-weight lbs, cubic inches-weight lbs.

```python
plt.subplots(3,2, figsize=(18,12))

plt.subplot(3,2,1)
sns.lineplot(x="hp" , y ="mpg" , data =df, color = "Orange")
plt.title("HP Effect on MPG")

plt.subplot(3,2,2)
sns.lineplot(x="hp" , y ="time-to-60" , data =df, color = "Orange")
plt.title("HP Effect on time-to-60")

plt.subplot(3,2,3)
sns.lineplot(x="mpg" , y ="time-to-60" , data =df, color = "Orange")
plt.title("MPG Effect on time-to-60")

plt.subplot(3,2,4)
sns.lineplot(x="hp" , y ="weightlbs" , data =df, color = "Orange")
plt.title("HP Effect on weight lbs")

plt.subplot(3,2,5)
sns.lineplot(x="mpg" , y ="weightlbs" , data =df, color = "Orange")
plt.title("MPG Effect on weight lbs")

plt.subplot(3,2,6)
sns.lineplot(x="cubicinches" , y ="weightlbs" , data =df, color = "Orange")
plt.title("cubic inches Effect on weight lbs")


plt.tight_layout()
plt.show()
```
